```python
import folium as f
import gpxpy as gx
import numpy as np
import pandas as pd
import polars as pl
from tqdm.notebook import trange, tqdm
import multiprocessing as mp
```

```python
myGpx = gx.parse(open("Escursione 2.gpx"))
myGpx2 = gx.parse(open("Escursione 3.gpx"))
```

```python
myGpx.tracks[0].get_elevation_extremes()
myGpx2.tracks[0].get_elevation_extremes()
```

```python
track = myGpx.tracks[0]
segment = track.segments[0]
# importing into Pandas
data = []
segment_len = segment.length_3d()

for point_idx, point in tqdm(enumerate(segment.points)):
    data.append([point.longitude, point.latitude, point.elevation, point.time, segment.get_speed(point_idx)])
    columns = ["Longitude", "Latitude", "Elevation", "Time", "Speed"]
    gpx_df = pd.DataFrame(data, columns=columns)
```

```python
points = []
for track in tqdm(myGpx.tracks):
    for segment in tqdm(track.segments):
        for point in tqdm(segment.points):
            points.append(tuple([point.latitude, point.longitude]))
```
